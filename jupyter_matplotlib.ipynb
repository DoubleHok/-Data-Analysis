## 完善原始折线图（辅助显示层）

```python
# 需求：画出头城市11点到12带你一小时内每分钟的温度变化折线图，温度范围在15度-18度
# 1、准备数据 x,y
import random
x=range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_Beijing = [random.uniform(1,3) for i in x]
# 2、生成画布
plt.figure(figsize = (20,8),dpi = 100)
# 3、绘制图像
plt.plot(x,y_shanghai,color = "r",linestyle = ":",label = "shanghai")
plt.plot(x,y_Beijing,color = "b",linestyle = "-.",label = "beijing")
# 修改刻度
# 准备x的刻度说明
x_label = ["11:{}".format(i) for i in x]
plt.xticks(range(0,61,5),x_label[::5])
plt.yticks(range(0,40,5))
# 添加网格显示
plt.grid(linestyle='--',alpha=0.5)
# 添加图例
plt.legend(loc = "upper left")
# 添加描述信息
plt.xlabel("Time variation")
plt.ylabel("Temperature variation")
plt.title("A graphic showing the temperature change between 11:00 and 12:00 noon")
# 4、显示图像
plt.show()
```

```python
# 需求：画出头城市11点到12带你一小时内每分钟的温度变化折线图，温度范围在15度-18度
# 1、准备数据 x,y
import random
x=range(60)
y_shanghai = [random.uniform(15,18) for i in x]
y_Beijing = [random.uniform(1,3) for i in x]
# 2、生成画布
# plt.figure(figsize = (20,8),dpi = 100)
figure,axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)
# 3、绘制图像
axes[0].plot(x,y_shanghai,color = "r",linestyle = ":",label = "shanghai")
axes[1].plot(x,y_Beijing,color = "b",linestyle = "-.",label = "beijing")
# 修改刻度
# 准备x的刻度说明
x_label = ["11:{}".format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label)
axes[0].set_yticks(range(0,40,5))
axes[1].set_xticks(x[::5])
axes[1].set_yticklabels(x_label)
axes[1].set_yticks(range(0,40,5))
# 添加网格显示
axes[0].grid(linestyle='--',alpha=1)# alpha 设置网格深浅度
axes[1].grid(linestyle='--',alpha=0.5)# alpha 设置网格深浅度
# 添加图例
axes[0].legend(loc = "upper left")
axes[1].legend(loc = "upper left")
# 添加描述信息
axes[0].set_xlabel("Time variation")
axes[0].set_ylabel("Temperature variation")
axes[0].set_title("A graphic showing the temperature change between 11:00 and 12:00 noon")
axes[1].set_xlabel("Time variation")
axes[1].set_ylabel("Temperature variation")
axes[1].set_title("A graphic showing the temperature change between 11:00 and 12:00 noon")
# 4、显示图像
plt.show()
```
