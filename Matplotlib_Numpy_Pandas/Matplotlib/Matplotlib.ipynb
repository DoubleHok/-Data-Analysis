###  调整X或者Y轴上的刻度

```python
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(10,5),dpi=200)
x = range(2,26,2) # 数据在x轴的位置，是一个可迭代对象
y = [15,13,14.5,17,20,25,26,26,24,22,18,15] # 数据在Y轴位置，是一个可迭代对象
plt.plot(x,y)
plt.xticks(x[::2]) # 当刻度太密集时候使用列表的步长（间隔取值）来解决，matplotlib会自动报哪个我们应对
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
plt.rc('font',family='SimHei',size=13) # 显示中文字体
plt.figure(figsize=(20,8),dpi=200)
x=range(120)
random.seed(10) # 设置随机种子，让不同时候随机得到的结果都一样
y=[random.uniform(20,35) for i in range(120)]
plt.plot(x,y)
_x_ticks = ["10点{}分".format(i) for i in x if i<60]
_x_ticks += ["11点{}分".format(i-60) for i in x if i>60]
plt.xticks(x[::5],_x_ticks[::5],rotation=90) # 让列表x中的数据和_x_ticks上的数据都传入，最终会在x轴上一一对应的显示出来
# 两组数据的长度必须一样，否则不能完全覆盖整个轴
# 使用列表的切片，每隔5个选一个数据进行展示
#  为了让字符串不会覆盖，使用rotation选项，让字符串旋转90°显示
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
plt.rc('font',family='SimHei',size=14)
plt.figure(figsize=(20,8),dpi=200)
x = range(120)
random.seed(2)# 设置随机种子，让不同时候随机得到的结果都一样。括号中的数字可以任意（表示一种随即状态）
y = [random.uniform(20,35) for i in range(120)]
plt.plot(x,y)
_x_ticks = ["10点{}分".format(i) for i in x if i<60]
_x_ticks += ["11点{}分".format(i-60) for i in x if i> 60]
plt.xticks(x[::5],_x_ticks[::5],rotation=90)
plt.xlabel("时间") # 设置X轴的label
plt.ylabel("温度/°C")# 设置Y轴的label
plt.title("10点到12点每分钟温度随时间变化情况") # 设置title
plt.show()
```

### 练习

```python
import matplotlib.pyplot as plt
import random
fig = plt.figure(figsize=(20,8),dpi=200)
plt.xlabel("年龄")
plt.ylabel("男(女)朋友/个")
plt.title("11岁-30岁交往的男(女)朋友个数分析")
a = range(11,31,1)
# y = [1,2,3,4,2,3,4,5,5,6,6,4,4,3,3,1,2,3,4,5]
random.seed(2)
b1 = [random.randint(1,4) for i in range(20)]
random.seed(3)
b2 = [random.randint(1,3) for i in range(20)]
plt.plot(a,b1,color='r',linestyle=':',linewidth=5,label="自己",alpha=0.5)
plt.plot(a,b2,color='g',linestyle='-.',linewidth=8,label="同桌",alpha=0.8)
# 图例的位置
plt.legend(bbox_to_anchor=(1,1),#图例边界框起始位置
                 loc="upper right",#图例的位置
                 ncol=1,#列数
                 mode="None",#当值设置为“expend”时，图例会水平扩展至整个坐标轴区域
                 borderaxespad=0,#坐标轴和图例边界之间的间距
                 title="主人公",#图例标题
                 shadow=True,#是否为线框添加阴影
                 fancybox=True)#线框圆角处理参数
_xticks_ = ["{}岁".format(i) for i in a ]
plt.xticks(x[::1],_xticks_[::1],rotation=45)
plt.savefig('./11岁-30岁交往的男(女)朋友个数分析.png')
plt.show()
```

### 饼状图

```python
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
#各部分标签
label_list = ["第一部分","第二部分","第三部分"]
#各部分大小
size = [55,35,10]
#各部分颜色
color = ["red","green","blue"]
#各部分突出值
explode = [0,0.05,0]
"""
绘制饼状图
explode:设置各部分突出值
label:设置各部分标签
autopct:设置圆里边文本
shadow:设置是否有阴影
startangle:起始角度，默认从0开始逆时针转
pctdistance:设置圆内文本距离圆心距离
radius  :控制饼图半径，默认值为1；
返回值
l_text:圆内部文本，matplotlib.text.Text object
p_text:圆外部文本
"""
pacthes,l_text,p_text = plt.pie(
    size,
    explode=explode,
    labels=label_list,
    labeldistance=1.1,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90,
    pctdistance=0.6)
plt.axis("equal")
#添加图例
plt.legend(loc="upper right",fontsize=10,bbox_to_anchor=(1.1,1.05),borderaxespad=0.3)
# loc =  'upper right' 位于右上角
# bbox_to_anchor=[0.5, 0.5] # 外边距 上边 右边
# ncol=2 分两列
# borderaxespad = 0.3图例的内边距
plt.savefig("./饼图01.png",dpi=200,bbox_inches='tight')  
plt.show()
```
